```python
import pandas as pd
import random
from datetime import datetime, timedelta

# PARAMS
num_rows = 15000   # 🔥 SHU YER O'ZGARDI

regions = {
    "US": ["USA"],
    "Europe": ["Germany", "France", "Italy"],
    "Asia": ["Japan", "India"],
    "South America": ["Brazil", "Argentina"],
    "China": ["China"]
}

cities = ["Berlin","Munich","Paris","Rome","Tokyo","Delhi","Sao Paulo","Beijing","New York"]

data = []

for _ in range(num_rows):
    # Random start date (2020–2026)
    start_date = datetime(2020,1,1) + timedelta(days=random.randint(0, 2500))
    stay_days = random.randint(1, 60)
    end_date = start_date + timedelta(days=stay_days)

    year = start_date.year
    if year > 2026:
        year = random.randint(2020, 2026)

    # Region / country
    region = random.choice(list(regions.keys()))
    country = random.choice(regions[region])
    city = random.choice(cities)

    # Rent
    avg_rent = random.randint(800, 2000)
    max_rent = avg_rent + random.randint(300, 1200)
    min_rent = avg_rent - random.randint(100, 400)

    # Days to rent
    days_to_rent = random.randint(1, 30)

    # Area (kv.m)
    area = random.randint(20, 150)

    # Daily rent
    daily_rent = round(avg_rent / 30, 2)

    # Income
    income = avg_rent * stay_days

    data.append([
        year, region, country, city,
        start_date, end_date,
        stay_days,
        avg_rent, max_rent, min_rent,
        days_to_rent,
        area,
        daily_rent,
        income
    ])

# DataFrame
columns = [
    "Year","Region","Country","City",
    "Start_Date","End_Date",
    "Stay_Days",
    "Avg_Rent","Max_Rent","Min_Rent",
    "Days_To_Rent",
    "Area_sqm",
    "Daily_Rent",
    "Rental_Income"
]

df = pd.DataFrame(data, columns=columns)

# KPI lar
df["Avg_Stay_Days"] = df["Stay_Days"].mean()
df["Max_Stay_Days"] = df["Stay_Days"].max()
df["Min_Stay_Days"] = df["Stay_Days"].min()

# SAVE
df.to_excel("real_estate_15000_rows.xlsx", index=False)

print("✅ 15000 row dataset tayyor!")
```
